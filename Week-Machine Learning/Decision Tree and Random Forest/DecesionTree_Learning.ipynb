```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading Data
df = pd.read_csv('golf_weather_dataset.csv')
df
```

```python
# Data Preprocessing
df.info()
```

```python
df.describe(include='all')
```

```python
df.head()
```

```python
# Drop the unnamed column
df_1 = df.drop(['Unnamed: 0'], axis=1)
```

```python
df_1
```

```python
# Create dummies for categorical variables except Play
y = df_1['Play']
x = df_1.drop(['Play'], axis=1)
```

```python
print(y)
print(x)
```

```python
x_dummies_vals = pd.get_dummies(x, columns=[['Windy', 'Humidity', 'Temperature', 'Outlook']])
x_dummies_vals
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

x_train, x_test, y_trian, y_test = train_test_split(x_dummies_vals, y, test_size=0.2)
```

```python
tree_obj = DecisionTreeClassifier()
model = tree_obj.fit(x_train, y_trian)
```

```python
# Predict

y_hat  = model.predict(x_test)
```

```python
# Model Accuracy
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(x_test, y_test))
```
