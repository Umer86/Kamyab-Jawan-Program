**Generally, lambda function is used with other functions**, *which are called filter,map and reduce function*.  


```python
"""lambda with filter: to filter values from some sequence
like (list, set, dictionary) on the basis of some conditions, 
than we use filter method in conjunction with lambda function """ 
```

```python
l1 = [1,2, 9, 50, 6,30]
odd_values = list(filter(lambda x: (x%2 != 0), l1))
```

```python
odd_values
```

```python
 l1 = [1,2, 9, 50, 6,30]
even_values = list(filter(lambda x: (x%2 == 0), l1))
```

```python
even_values
```
